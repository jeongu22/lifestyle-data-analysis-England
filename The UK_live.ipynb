```python
import numpy as np
import pandas as pd
import matplotlib as mlp
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data1 = pd.read_csv('british.csv')
data1
```

```python
data2 = pd.read_csv('british2.csv')
data2
```

```python
data = pd.concat([data1,data2], axis = 0, ignore_index = True)
data
```

```python
# AI참조 - (f.g 시간을 09:45처럼 표현하고 싶을 때 )

# data['cellphone_usage'] = (data['cellphone_usage'].dt.components['hours'].astype(str).str.zfill(2) + ':' + data['cellphone_usage'].dt.components['minutes'].astype(str).str.zfill(2))
```

```python
data['date_dt'] = pd.to_datetime(data['date'], format = '%Y-%m-%d')
data['cellphone_usage'] = pd.to_timedelta(data['cellphone'])
```

```python
order = ['Mon','Tue','Wed','Thu','Fri','Sat','Sun']
data['day'] = pd.Categorical(data['day'], categories = order, ordered = True)
data['academy'] = data['academy'].replace({0:'결석',1:'출석'})
data['academy'] = data['academy'].fillna('휴일')
```

```python
data.info()
```

```python
data['country'].unique()
```

```python
day_fee = data.groupby(['day'])['food_fee'].mean()
day_fee
```

```python
#연습

x= day_fee.index
y= day_fee.values

fig,ax = plt.subplots()
ax.bar(x,y)
```

-또한 운동하는 날에 박스 크기가 비교적 작은 것으로 보아, 운동하는 날이 비교적 규칙적인 휴대폰 사용시간을 나타내기도 한다.
